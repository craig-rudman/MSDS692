## II. Problem Statement

Federal workforce reductions in early 2025 had documented effects on National Weather Service staffing through probationary terminations and waves of deferred resignations. Contemporaneous press reporting characterized the impact in urgent terms, yet such accounts, while not inaccurate, lacked quantitative grounding. The public was left to weigh alarming headlines against official reassurances, with no independent empirical basis for judgment.

This situation reflects a broader failure of the oversight mechanisms that normally earn and sustain public trust. In a functioning democratic system, government agencies conduct internal oversight, inspectors general audit performance, Congress provides external accountability, and journalism validates or challenges official accounts. When those mechanisms are weakened simultaneously, as they were in 2025 with inspector general firings, diminished congressional oversight, and a media environment in which press credibility is itself contested, citizens lose the institutional anchors that ordinarily allow them to evaluate government performance.

This project asks whether publicly available federal data can substitute, at least partially, for those failed mechanisms. Specifically, using NWS warning event records and Local Storm Report data, can a citizen analyst empirically evaluate whether the 2025 staffing reductions measurably degraded the quality of weather warning services received by the American public? The analysis is observational and the data imperfect, but the method is transparent and reproducible, auditable by any reader with the same public data. That epistemic property is the core contribution: not just a finding, but a demonstration that independent public inquiry is possible.
